In this notebook we will open a Relion .star file containing particle data, and plot some of the features

```python
from StarIO import read_star, write_star # Star IO functions from here: https://github.com/cschlick/StarIO
import pandas as pd 
%matplotlib inline 
```

```python
df_dict = read_star("examples/run_it013_data.star")
```

```python
df = df_dict["data"]
print(df.columns)
```

1. Check the number of particles

```python
len(df)
```

2. Histogram the defocus range of the particles

```python
ax = df["rlnDefocusV"].hist(bins=50)
ax.set_xlabel("Defocus")
```

3. Histogram the CTFind Max Resolution estimate

```python
ax = df["rlnCtfMaxResolution"].hist(bins=50)
ax.set_xlabel("Max Resolution (angstroms)")
```

4. Plot both to see a correlation

```python
ax = df.plot.scatter(x="rlnCtfMaxResolution",y="rlnDefocusU")
```

```python
ax = df.plot.scatter(x="rlnCoordinateX",y="rlnCoordinateY")
```

```python
ax = df.plot.scatter(x="rlnMaxValueProbDistribution",y="rlnLogLikeliContribution")
```
