In this notebook we plot the class distribution from a Relion classification

```python
from StarIO import read_star, write_star # Star IO functions from here: https://github.com/cschlick/StarIO
import pandas as pd 
%matplotlib inline 
```

```python
df_dict = read_star("examples/run_it008_model.star")
```

```python
df = df_dict["model_classes"]
df["Class Number"] = [x for x in range(1,11)] # add class number attribute
```

```python
ax = df.plot.bar(x="Class Number",y="rlnClassDistribution",legend=False)
ax.set_ylabel("Class Distribution")
```

```python
df.sort_values("rlnClassDistribution",ascending=False,inplace=True)
df.plot.bar(x="Class Number",y="rlnClassDistribution",legend=False)
ax.set_ylabel("Class Distribution")
```
